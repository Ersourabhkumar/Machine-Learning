```python
## SUPPORT VECTOR MACHINE(SVM) -->svm is a supervised ml algorithm that is used for classification. it is perticulerly effective in dealing with complex and high diamentional data set
# the fundamental principle of svm is to find an optimal hyper plane that maximaly saprates different classes in the input space.

# HOW SVM WORKS---> svm requres labled training data consisting of input features and corrosponding class lables. each data point is representad as a feature vector
#  where each feature describe a perticular charectorstics of the data point. the data point should be preprocessed and scalled to ensure the features are on similer scales.
#  typically 0 and 1.

## HYPERPLANE AND MARGIN----> svm aims to find a hyperplane that best saprets the different classes in the feature space. in a binory classsification problem the hyperplane is a line
# in a 2d space or a hyperplane in a  higher diamentional space svm seeks to maximize the margin which is the distance between the hyperplane
# and nearest data points from each class the points on the margin are known as support vectors as they play a crocial role in defining the decision boundry

## LINEAR SVM---> the linear svm finds a linear hyperplane that saprets the classes the goal is to find the hyperplane that maximize the margin while minimizing the missclassification of training examples.
# mathematicaly this can be formulated as an optmization problem with the objective of minimizing the weigts of the hyperplane subject to the constraint that .all training examples
# lie on the currect side of the hyperplane

## NON LINEAR SVM---> in cases where the data is not linearly saprated svm uses the technique that called the kernel trick . the kernel trick meps the original input space into a
# heigher diamentional feaure space where the datapoints can be linearly sapratable

## TRAINING ---> svm training involves finding the optimal hyperplane or decision boundry that saprates the classes the optimization problem is typically solved using mathods such as
# quadratic programing or sequential minimal optimization . this process involves solving for the weights of the hyperplane and the bias term which defines the decision boundry
# the objective is to minimize the regularization term while esuring that the training examples are correctly classified

## PRIDCITION---> once the svm model is trained it can be used to predict the class lables of new unseen data points the model computes the distance from the test points to the decesion boundry
# the predicted class lables is determined based on which side of decesion boundry the point live . the decision function can also provide a confidence score or probability estimate
# for the prediction

## ADVANTAGES---> it effective in heigher diamentional spaces it is rowbest against overfitting due to marginn maximization principle it is vercitile through the use of different kernel
# functions . it can handle both linear and non linear classification task
```

```python
from sklearn import svm
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Generate synthetic classification data
x,y=make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,random_state=42)

# Spilt the data into training and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# Linear Kernel
linear_svm=svm.SVC(kernel='linear')
linear_svm.fit(x_train,y_train)
linear_predictions =linear_svm.predict(x_test)
linear_accuracy=accuracy_score(y_test,linear_predictions)
print("Linear Kernel Accuracy:",linear_accuracy)
```

```python
# Polynomial Kernel
poly_svc=svm.SVC(kernel='poly',degree=3) #degree = 3 ==> x**0 + x**1 + x**2
poly_svc.fit(x_train,y_train)
poly_predictions=poly_svc.predict(x_test)
poly_accuracy=accuracy_score(y_test,poly_predictions)
print("Polynomial Kernel Accuracy:",poly_accuracy)
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("cars - cars.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb=LabelEncoder()
```

```python
df["name"]=lb.fit_transform(df["name"])
df["fuel"]=lb.fit_transform(df["fuel"])
df["seller_type"]=lb.fit_transform(df["seller_type"])
df["transmission"]=lb.fit_transform(df["transmission"])
df["owner"]=lb.fit_transform(df["owner"])
```

```python
df.head(2)
```

```python
x=df.drop(columns=["selling_price"])
y=df["selling_price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
# y=mx+c
```

```python
param_space={"copy_X":[True,False],"fit_intercept":[True,False],
            "n_jobs":[1,5,10,15,None],"positive":[True,False]}
```

```python
grid_search=GridSearchCV(lr,param_space,cv=3)
```

```python
grid_search.fit(x_train,y_train)
```

```python
y_pred_new=grid_search.predict(x_test)
```

```python
r2_score(y_test,y_pred_new)
```

```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("covid_toy - covid_toy.csv")
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df["gender"]=df["gender"].map({"Male":0,"Female":1})
df["cough"]=df["cough"].map({"Mild":0,"Strong":1})
df["city"]=df["city"].map({"Kolkata":0,"bangalore":1,"Delhi":2,"Mumbai":3})
df["has_covid"]=df["has_covid"].map({"Yes":0,"No":1})
```

```python
df.head(3)
```

```python
x=df.drop(columns=["has_covid"])
y=df["has_covid"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
np.round(x_train.describe(),2)
```

```python
# Now we  will  apply standardization
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train) #fit means  learn the parameter
#tranform means apply on that data
```

```python
x_train_new=pd.DataFrame(x_train_sc,columns=x.columns)
```

```python
np.round(x_train_new.describe(),2)
```

```python
import numpy as np
import pandas  as pd
```

```python
df=pd.read_csv("crocodile_dataset - crocodile_dataset.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.drop(columns=["Notes","Observation ID","Date of Observation"])
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns=["Conservation Status"])
y=df["Conservation Status"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df.dtypes
```

```python
x_train["Age Class"].value_counts
```

```python
x_train.head(2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler ,OrdinalEncoder ,OneHotEncoder
```

```python
transformer=ColumnTransformer(transformers=[
    ("scalling",StandardScaler(),["Observed Length (m)","Observed Weight (kg)"]),
    ("Ordinal",OrdinalEncoder(categories=[["Adult","Juvenile","Subadult",'Hatchling']]),["Age Class"]),
    ("Onehot",OneHotEncoder(sparse_output=False, drop="first"),["Common Name","Scientific Name","Family","Genus","Sex","Country/Region","Habitat Type","Observer Name"])
    
],remainder="passthrough")
```

```python
x_train_tf=transformer.fit_transform(x_train)
x_train_tf.shape
```
